```python
for i in range(0, img.shape[0], p):
    for j in range(0, img.shape[1], p):
        ls.append(img[i:i+p, j:j+p])
        
```

# Esercizio 1

```python
img = cv2.imread("media/Lautaro.JPG")

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap = "gray")
```

```python
imgGray_binary = np.empty((img.shape[0], img.shape[1]), dtype=object)  
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        imgGray_binary[i, j] = format(img[i, j], '08b')
```

```python
imgGray_binary
```

```python
def bit_plane(img, n):
    h = img.shape[0]
    w = img.shape[1]
    bit_array = np.zeros((h, w), dtype = int)
    for i in range(h):
        for j in range(w):
            bit_array[i,j] = imgGray_binary[i,j][n]
    plt.imshow(bit_array, cmap='gray')
    plt.show()
            
```

```python
bit_plane(img, 2)
```

```python
t = int(input("Choose a value for T. Either 0 or 1"))      
```

```python
for n in range(8):
    if t == 0:
        if n%2 != 0:
            for i in range(img.shape[0]):
                for j in range(img.shape[1]):
                    binary_string = imgGray_binary[i, j]  # Supponendo che sia una stringa binaria
                    binary_list = list(binary_string)    # Converti in lista
                    binary_list[n] = "0"                 # Modifica il carattere all'indice n
                    imgGray_binary[i, j] = ''.join(binary_list)  # Riconverti in stringa
    if t == 1:
        if n%2 == 0:
            for i in range(img.shape[0]):
                for j in range(img.shape[1]):
                    binary_string = imgGray_binary[i, j]  # Supponendo che sia una stringa binaria
                    binary_list = list(binary_string)    # Converti in lista
                    binary_list[n] = "0"                 # Modifica il carattere all'indice n
                    imgGray_binary[i, j] = ''.join(binary_list)  # Riconverti in stringa
    bit_plane(img, n)
```
