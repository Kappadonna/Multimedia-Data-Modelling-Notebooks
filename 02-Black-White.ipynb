```python
import numpy as np
import cv2
from  PIL import Image, ImageEnhance 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('media/abarth.png')

# Convert the image from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('on')
plt.show()
```

```python
pixel = image[250, 250]
```

```python
print("B: ", pixel[0])
print("G: ", pixel[1])
print("R: ", pixel[2])
```

```python
p =100

for i in range(494):
    for j in range(700):
        pixel = image[i, j]
        if pixel[0] and pixel[1] and pixel[2] < p:
            pixel[0:2] = 0
        else:
            pixel[0:2] = 255
```
