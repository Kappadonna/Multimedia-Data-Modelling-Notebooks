```python
import cv2

import matplotlib.pyplot as plt

import numpy as np
```

```python
image = cv2.imread('media/Zante.JPG')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis('on')
plt.show()
```

```python
def zoom_img(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    h = img.shape[0]
    w = img.shape[1]
    zoom_image = np.zeros((h*2, w*2), dtype = int)
    for i in range(h):
        for j in range(w):
            zoom_image[2*i, 2*j] = img[i, j]
            zoom_image[2*i + 1, 2*j + 1] = img[i, j]
            zoom_image[2*i + 1, 2*j] = img[i, j]
            zoom_image[2*i, 2*j + 1] = img[i, j]
    plt.imshow(zoom_image, cmap = "gray")
```

```python
def zoom_in(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    h = img.shape[0]
    w = img.shape[1]
    zoom_pic = np.zeros((h, w), dtype = int)
    for i in range(h//2):
        for j in range(w//2):
            zoom_pic[2*i, 2*j] = img[i, j]
            zoom_pic[2*i + 1, 2*j + 1] = img[i, j]
            zoom_pic[2*i + 1, 2*j] = img[i, j]
            zoom_pic[2*i, 2*j + 1] = img[i, j]
    plt.imshow(zoom_pic, cmap = "gray")
```

```python
def zoomed_img(pic):
    h = pic.shape[0]
    w = pic.shape[1]
    c = pic.shape[2]
    pic_zoomed = np.zeros((h, w, c), dtype = np.uint8)
    for x in range(h//2):
        for y in range(w//2):
            for z in range(c):
                pic_zoomed[2*x, 2*y, z] = pic[x, y, z]
                pic_zoomed[2*x + 1 , 2*y + 1, z] = pic[x, y, z]
                pic_zoomed[2*x + 1, 2*y, z] = pic[x, y, z]
                pic_zoomed[2*x, 2*y +1 , z] = pic[x, y, z]
    plt.imshow(pic_zoomed)
```

```python
zoom_img(image)
```

```python
zoomed_img(image)
```
