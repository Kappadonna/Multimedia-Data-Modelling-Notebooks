```python
def zoom_in(img):
    h = img.shape[0]
    w = img.shape[1]
    zoom_pic = np.zeros((h, w), dtype = int)
    for i in range(h//2):
        for j in range(w//2):
            zoom_pic[2*i, 2*j] = img[i, j]
            zoom_pic[2*i + 1, 2*j + 1] = img[i, j]
            zoom_pic[2*i + 1, 2*j] = img[i, j]
            zoom_pic[2*i, 2*j + 1] = img[i, j]
    plt.imshow(zoom_pic, cmap = "gray")
```

```python
zoom_in(imgGray)
```

```python
def zoom_img(img):
    h = img.shape[0]
    w = img.shape[1]
    zoom_image = np.zeros((h*2, w*2), dtype = int)
    for i in range(h):
        for j in range(w):
            zoom_image[2*i, 2*j] = img[i, j]
            zoom_image[2*i + 1, 2*j + 1] = img[i, j]
            zoom_image[2*i + 1, 2*j] = img[i, j]
            zoom_image[2*i, 2*j + 1] = img[i, j]
    plt.imshow(zoom_image, cmap = "gray")
```

```python
zoom_img(imgGray)
```

```python

zoom_picture = np.zeros((1280, 720, 3), dtype=np.uint8)

for i in range(640):
    for j in range(360):
        for c in range(3):  # Iterate over each channel (0: Red, 1: Green, 2: Blue)
            zoom_picture[2*i, 2*j, c] = image[i, j, c]
            zoom_picture[2*i + 1, 2*j + 1, c] = image[i, j, c]
            zoom_picture[2*i + 1, 2*j, c] = image[i, j, c]
            zoom_picture[2*i, 2*j + 1, c] = image[i, j, c]


plt.imshow(zoom_picture)
```

```python
def zoomed_img(pic):
    h = pic.shape[0]
    w = pic.shape[1]
    c = pic.shape[2]
    pic_zoomed = np.zeros((h, w, c), dtype = np.uint8)
    for x in range(h//2):
        for y in range(w//2):
            for z in range(c):
                pic_zoomed[2*x, 2*y, z] = pic[x, y, z]
                pic_zoomed[2*x + 1 , 2*y + 1, z] = pic[x, y, z]
                pic_zoomed[2*x + 1, 2*y, z] = pic[x, y, z]
                pic_zoomed[2*x, 2*y +1 , z] = pic[x, y, z]
    plt.imshow(pic_zoomed)
    return(pic_zoomed)
```

```python
zoomed_img(image)
```
