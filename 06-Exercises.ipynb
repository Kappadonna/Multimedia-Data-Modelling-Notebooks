```python
import numpy as np

import random
```

```python
m = int(input("Insert number of rows:"))

n = int(input("Insert number of columns:"))

image = np.random.randint(0, 255, size=(m, n, 3))
```

```python
# Define a function that swaps even-numbered columns with odd-numbered columns and viceversa
def swap_col(img):
    swap_img = img.copy()
    for i in range(m):
        for j in range(0, n-1, 2):
            swap_img[i,j] = img[i, j+1]
            swap_img[i,j+1] = img[i, j]
        
    return swap_img
```

```python
swap_col(image)
```

```python
image
```

```python
#Define a function that exchanges even rows with odd rows and vice versa

def swap_row(img):
    swap_img = img.copy()
    for i in range(0, m-1, 2):
        for j in range(n):
            swap_img[i,j] = img[i+1, j]
            swap_img[i+1,j] = img[i, j]
        
    return swap_img
```

```python
swap_row(image)
```

```python
image
```

```python
X = 229

def set_to_zero(img):
    X = int(input("Select a random number X, if one of the pixel's channels has a value equal to X, all the channels are set to 0")) 
    for x in range(m):
        for y in range(n):
            for z in range(3):
                if img[x, y, z] == X:
                    img[x, y] = 0
    return image
```
