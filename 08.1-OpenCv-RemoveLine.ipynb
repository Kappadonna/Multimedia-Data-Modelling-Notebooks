```python
for i in range(img.shape[0]):
    if i != 0 and i != img.shape[0]-1:
        img[i, i] = img[i, i-1]/2 + img[i, i+1]/ 2
    if i == 0:
        img[i, i] = img[i, i+1]
    if i == img.shape[0]-1:
        img[i, i] = img[i, i-1]
```

### Interpolation with 4 elements

```python
for i in range(img.shape[0]):
    if i != 0 and i != img.shape[0]-1:
        img[i, i] = img[i-1, i]/4 + img[i+1, i]/4 + img[i, i-1]/4 + img[i, i+1]/4
    if i == 0:
        img[i, i] = img[i, i+1]/2 + img[i+1, i]/2
    if i == img.shape[0]-1:
        img[i, i] = img[i, i-1]/2 + img[i-1, i]/2
```

```python
plt.imshow(img)
```

# Fastest way

Do not work on the image, but on a copy of it

```python
imgBGR = cv2.imread("Introduction to OpenCV/lena.png")

img =  cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
imgLine = img.copy()
```

```python
blue = (0, 0, 255)

cv2.line(imgLine,(0,0),(img.shape[0], img.shape[1]), blue, thickness = 1)

plt.imshow(imgLine)
```
