```python
import numpy as np
```

```python
m = 11

n = 11

image = np.random.randint(0, 255, size=(m, n, 3))

image
```

```python
image.shape
```

```python
R = image[:, :, 0]

R
```

```python
G = image[:, :, 1]

G
```

```python
B = image[:, :, 2]

B
```

```python
bayer_array = np.zeros((m, n), dtype = int)

bayer_array
```

```python
for i in range(m):
    for j in range(n):
        if i%2 == 0 and j%2 == 0:
            bayer_array[i, j] = R[i, j]
        elif i%2 != 0 and j%2 !=0:
            bayer_array[i, j] = B[i, j]
        else:
            bayer_array[i, j] = G[i, j]
```

```python
bayer_array
```

```python
I2 = array = np.full((m, n, 3), -1)

I2
```

```python
R2 = I2[:, :, 0].copy()

G2 = I2[:, :, 1].copy()

B2 = I2[:, :, 2].copy()
```

```python
for i in range(m):
    for j in range(n):
        if i%2 == 0 and j%2 == 0:
            R2[i, j] = bayer_array[i, j]
        elif i%2 != 0 and j%2 !=0:
            B2[i, j] = bayer_array[i, j]
        else:
            G2[i, j] = bayer_array[i, j]
            
```

```python
R2
```

```python
for i in range(m):
    for j in range(n):
        if i % 2 == 0 and j% 2 ==0:
            if i+1 < m and j+1 < n:
                R2[i+1, 1+j] = R2[i, j]
            if j+1 < n:
                R2[i, j+1] = R2[i, j]
            if i+1 < m:
                R2[i+1, j] = R2[i, j]
        
```

```python
R2
```

```python
B2
```

```python
for i in range(m):
    for j in range(n):
        if i % 2 != 0 and j % 2 != 0:
            B2[i-1, j-1] = B2[i, j]
            B2[i, j-1] = B2[i, j]
            B2[i-1, j] = B2[i, j]    
        
    
for i in range(m):
    for j in range(n):         
        if j == n-1:
            B2[i, j] = B2[i, j-1]  
        if i == m-1:
            B2[i, j] = B2[i-1, j]
```

```python
B2
```

```python
G2
```

```python
for i in range(m):
    for j in range(1, n, 2):
        if i % 2 == 0:
            G2[i, j - 1] = G2[i, j]
            if j == n-2:
                G2[i, j+1] = G2[i, j]
```

```python
G2
```

```python
for i in range(m):
    for j in range(0, n, 2):
        if i % 2 != 0 and j+1 < n:
            G2[i, j + 1] = G2[i, j]
```

```python
G2
```

```python
I2[:, :, 0] = R2

I2[:, :, 1] = B2

I2[:, :, 2] = G2

I2
```

```python
image
```

```python
mse = np.mean((image - I2) ** 2)

print(f"MSE: {mse}")
```

```python
max_pixel_value = 255.0

psnr = 10 * np.log10((max_pixel_value ** 2) / mse)

print(f"PSNR: {psnr} dB")
```
